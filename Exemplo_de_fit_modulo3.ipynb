## Custo

$$ E = \frac{1}{n} \sum_{i=0}^{n} (y_{i} - (ax_i^2 + bx_i +c))^2 $$

```
def custo(y_real,y_estimado):
  n = y_estimado.shape[0]
  E = (1/n)*sum((y_real-y_estimado)**2)
  return E
```

Derivada parcial com relao ao coeficiente a:

$$ \frac{dE}{da} = \frac{-2}{n} \sum x_i^2 (y_i - (ax_i^2 +bx_i +c)) $$


Derivada parcial com relao ao coeficiente b:

$$ \frac{dE}{db} = \frac{-2}{n} \sum x_i (y_i - (ax_i^2 +bx_i +c)) $$


Derivada parcial com relao ao coeficiente c:

$$ \frac{dE}{dc} = \frac{-2}{n} \sum (y_i - (ax_i^2 +bx_i +c)) $$



$$ a_i = a_0 - passo*\frac{dE}{da}  $$

$$ b_i = b_0 - passo*\frac{dE}{db}  $$

$$ c_i = c_0 - passo*\frac{dE}{dc}  $$

```
def grad(a,b,c,x,y,n):
  ga = (-2/n)*sum((x**2)*(y-(a*x**2)+b*x+c))
  gb = (-2/n)*sum((x)*(y-(a*x**2)+b*x+c))
  gc= (-2/n)*sum((y-(a*x**2)+b*x+c))

  return ga,gb,gc
```

```
a = 5
b =2
c = 2
```

```
def atualiza(a,b,c,x,y,n):
  ga,gb,gc = grad(a,b,c,x,y,n)

  a_novo = a - 0.00005*ga
  b_novo = b - 0.00001*gb 
  c_novo = c - 0.0009*gc

  return a_novo, b_novo, c_novo
```
